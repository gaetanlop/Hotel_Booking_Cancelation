# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import chi2,SelectKBest, mutual_info_classif
```

```python
df=pd.read_csv('hotel_bookings_EDA.csv')
df.head()
```

```python
df.shape
```

# Handle Missing Values

```python
df.isnull().sum()
```

Four features have missing values. We will go throw them one by one in order to decide what technique will we use to deal with them

### Let's have a look at the children feature

As there are only 4 rows with missing values, it will not affect our model if we drop them. Thus, we will drop these rows.

```python
df1=df[~(df.children.isnull())]
```

```python
df1.isnull().sum()
```

### Let's have a look at the company feature

```python
percentage_missing=df1.company.isnull().sum()/df1.shape[0]*100
print(f'Missing values represent {int(percentage_missing)} % of the feature company')
```

We will drop the column.

```python
df2=df1.drop('company',axis=1)
```

```python
df2.isnull().sum()
```

### Let's have a look at the agent feature

The feature Agent repressents the ID of the travel agency that made the booking. Company represents the entity that made the booking or responsible for paying the booking. These two features are the same. We will also drop these feature as it will not impact our model predictions.

```python
df3=df2.drop('agent',axis=1)
```

### Let's have a look at the country feature

```python
df3.country.value_counts()
```

```python
percentage_missing=df3.country.isnull().sum()/df3.shape[0]*100
print(f'Missing values represent {percentage_missing} % of the feature country')
```

Let's drop these values as it represents less than 1% of the total number of rows. It will not have any impact on our model if we drop these values.

```python
df4=df3[~(df3.country.isnull())]
```

```python
df4.isnull().sum()
```

# Feature Engineering

In this section, we are trying to find out if adding or removing features will be efficient for our model.

### Correlation matrix

Let's first convert categorical data to integer (using Label Encoder) in order to look at the correlations

```python
#find all categorical features
cats=df4.select_dtypes(include=['object']).columns
```

```python
le=LabelEncoder()
df_encoded=df4.copy()

#encoding all categorical variables
for cat in cats:
    df_encoded[cat] = le.fit_transform(df_encoded[cat])

# Creating the correlation matrix
plt.figure(figsize=(20,10))
correlation_full=df_encoded.corr()
sns.heatmap(correlation_full, annot=True)
```

.... CONCLUSION ....

```python
#Trying to understand the relaitonship between the independent features and the dependent feature.
correlation_cancel=correlation_full['is_canceled'].sort_values(ascending=False)
print(correlation_cancel)
```

Looking at this, it seems that 'deposit_type', 'lead_time','country', 'total_of_special_requests','required_car_parking_spaces' are the features that are the most correlated with the dependent variable. 

### Feature Generation

```python
#Creating a new feature using the country feature
df4['home_country']=df4.country.apply(lambda x: 1 if x=='PRT' else 0)
```

```python
df5=df4.drop('country',axis=1)
```

### Deal with Data Leakage

It seems that some informations in this dataset won't be availmable at the time we need to make the predictions. We need to delete these features.
'days_in_the_waiting_list' is an example of data leakage. Indeed, it represents  the number of days the booking was in the waiting list before it was confirmed to the customer. This information won't be available at the time of the prediction since we need to predict wether the booking will be canceled or not just after the reservation is made by the customer.

```python
df6=df5.drop('days_in_waiting_list',axis=1)
```

Let's also drop the feature arrival_date_year because I believe that it will not have any effects on the dependent variable.

```python
#Dropping the column arrival_date_year
df7=df6.drop('arrival_date_year',axis=1)

#Dropping the column assigned_room_type to prevent data leakage
df8=df7.drop('assigned_room_type',axis=1)

#Dropping the column booking_changes to prevent data leakage
df9=df8.drop('booking_changes',axis=1)
```

### Saving the new file

```python
df9.to_csv('hotel_bookings_cleaned.csv',index=False)
```

### Feature Selection

```python
df=pd.read_csv('hotel_bookings_cleaned.csv')
```

```python
df.head()
```

Before using SelectKBest for Feature Selection, we need to find the transform categorical variables into numerical types.

I will label encode the feature arrival date month and one hot encode the other features.

```python
#Using LabelEncoder of Sklearn to transforme the feature arrival_date_motnh wich has a high cardinality (12 months in a year)
le=LabelEncoder()
df_le=le.fit_transform(df.arrival_date_month)
df.arrival_date_month=df_le
```

```python
df=pd.get_dummies(df)
df.shape
```

### Let's do some Feature Selection using SelectKBest from Sklearn.

```python
# Univariate selection
X= df.drop('is_canceled',axis=1)
y=df.is_canceled
```

```python
X.head()
```

```python
y.head()
```

```python
# code for the SelectKBest algo: https://github.com/krishnaik06/Feature-Engineering-Live-sessions/blob/master/Feature%20Selection-Day%207.ipynb

# Apply SelectKBest algorithm
ordered_rank_features=SelectKBest(score_func=mutual_info_classif,k=52)
ordered_feature=ordered_rank_features.fit(X,y)

#Put the results in a DataFrame
dfscores=pd.DataFrame(ordered_feature.scores_,columns=["Score"])
dfcolumns=pd.DataFrame(X.columns)
features_rank=pd.concat([dfcolumns,dfscores],axis=1)

features_rank.columns=['Features','Score']
features_rank.sort_values('Score',ascending=False)
```

```python
#Keeping only the 44th first apartments
columns_keep=features_rank.sort_values('Score',ascending=False).Features.values[:44]

df1=df[columns_keep].copy()

#concatenating the new dataframe with the dependent variable
df2=pd.concat([df1,y],axis=1)
```

```python
df2.to_csv('hotel_bookings_cleaned_Kbest.csv',index=False)
```
