# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('hotel_bookings_cleaned.csv')
df.head()
```

# Exploratory Data Analysis

Let us explore the dependent variable.

```python
#Check if we are in the case of an imbalanced dataset.
sns.countplot(x='is_canceled',data=df)
```

We are clearly not in the case of an imbalanced dataset (60%-40% ratio).

### Exploring the relationship between the different categorical features and the dependent variable.

```python
df.info()
```

```python
#categorical variables
cat_var=['arrival_date_month','meal','country','market_segment','distribution_channel','reserved_room_type','deposit_type','customer_type']
```

```python
for cat in cat_var:
    plt.xticks(rotation=90)
    plt.title(cat)
    sns.countplot(x=cat,hue='is_canceled',data=df)
    plt.show()
```

* It seems that the cancelation rate is approximately the same for each month, for each meal type and for each room type.
* It seems that the cancelation rate for groups is higher.
* It seems that the cancelation rate for TA/TO distribution channel is higher.
* We should explore the variable deposit_type.
* It seems that the cancelation rate for transient, contract and group is higher.

Let's explore the variable deposit_type.

```python
df.groupby('is_canceled').deposit_type.value_counts()
```

#### We need to find a way to transform the column country.

```python
#let us create a new variable corresponding to whether or not the reservation was made by people living in Portugal.
df['home_country']=df.country.apply(lambda x: 1 if x=='PRT' else 0)

#drop the feature country
df.drop('country',axis=1,inplace=True)
```

```python
sns.countplot(x='home_country',hue='is_canceled',data=df)
```

It seems that the reservation rate is higher for people coming from Portugal.

### Exploring the relationship between the different numerical features and the dependent variable.

```python
df.columns
```

```python
num_var=['lead_time','stays_in_weekend_nights','stays_in_week_nights','adults',
         'children','babies','adr','required_car_parking_spaces','total_of_special_requests']
```

```python
for num in num_var:
    sns.barplot(x='is_canceled',y=num,data=df)
    plt.show()
```

Looking at these plots:
* The cancelation rate is higher for people with higher lead_time.
* The cancelation rate is lower for people with babies.
* The cancelation rate is lower for people who made less special requests.

### Let's explore how bookings price for each hotels evolve during the year

```python
price_month=df[['arrival_date_month','adr','hotel','adults','children','babies']].copy()
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]
price_month["arrival_date_month"] = pd.Categorical(price_month["arrival_date_month"], categories=months, ordered=True)


plt.figure(figsize=(12,9))
plt.title('price evolution during the month of the year')
sns.lineplot(x = "arrival_date_month", y="adr", hue="hotel", data=price_month)
```

### Just for curiosity, let's check out the number of nights people stay at the hotel

```python
nights=df[['hotel','stays_in_week_nights','stays_in_weekend_nights']].copy()
nights['full']=nights['stays_in_week_nights']+nights['stays_in_weekend_nights']
plt.figure(figsize=(15,8))
plt.title('number of nights people stay at hotal')
sns.countplot(x='full',data=nights,hue='hotel')
```

### Correlation

```python
corr=df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr,annot=True)
```

```python
#is_canceled correlation matrix
# code for the heatmap: https://www.kaggle.com/pmarcelino/comprehensive-data-exploration-with-python
k = 10 #number of variables for heatmap
cols = corr.nlargest(k, 'is_canceled')['is_canceled'].index
cm = np.corrcoef(df[cols].values.T)

plt.figure(figsize=(12,8))
sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, 
            yticklabels=cols.values, xticklabels=cols.values)
```

```python
# Let us create another feature for the family size
df['family_size']=df.adults+df.children+df.babies
```
